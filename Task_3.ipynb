##Segment wise analysis of Profit, Discount and Sales

```python
df['Segment'].value_counts()
```

```python
df_segment= df.groupby(['Segment'])[['Sales', 'Discount', 'Profit']].mean()
df_segment
```

```python
#1. sales 2. Discount 3. Profit
df_segment.plot.pie(subplots=True, 
                    autopct='%1.1f%%',
                    figsize=(15, 20),     # start angle 90°
                    shadow=True,
                    labels = df_segment.index)
plt.title('Segment wise analysis of Sale, Discount, profit')
```

##State wise analysis of Profit, Discount and Sales

```python
df['State'].value_counts().head(10)
```

```python
df_state= df.groupby(['State'])[['Sales', 'Discount', 'Profit']].mean()
df_state.head(5)
```

###State wise Profit Analysis

```python
df_state=df_state.sort_values('Profit')

df_state[['Profit']].plot(kind = 'bar', figsize = (15,4))
plt.title('State wise Profit Analysis', fontsize = 20)
plt.ylabel('Profit per Sate')
plt.xlabel('States')
plt.show()
```

###From above result are

####Vermont: Highest Profit

####Ohio: Lowest Profit

###State wise Sales Analysis

```python
df_state['Sales'].plot(kind='pie',
                        figsize = (15,15),
                        autopct='%1.1f%%',
                        startangle=90,     # start angle 90° (Africa)
                        shadow=True)
plt.title('State wise analysis of Sale',fontsize=20)
```

###From above results are

####Highest amount of sales= Wyoming(11.8%)

####Lowest amount of sales= South Dakota(0.8%)

###State wise Discount Analysis

```python
df_state['Discount'].plot(kind='bar',figsize=(20,5))
plt.title('State wise analysis of Discount', fontsize=20)
```

###From above results are

####Highest Discount= Illinois

##Quantity wise Analysis of the Profit, Sales and Discount

```python
df_quantity = df.groupby(['Quantity'])[['Sales', 'Discount', 'Profit']].mean()
df_quantity.head(5)
```

```python
#1. sales 2. Discount 3. Profit
df_quantity.plot.pie(subplots=True, 
                    autopct='%1.1f%%',
                    figsize=(25, 25),
                     pctdistance=0.69,
                    startangle=90,     # start angle 90° (Africa)
                    shadow=True,
                    labels = df_quantity.index)
plt.title('Quantity wise analysis of Sale, Discount, profit')
```

###From above
####13 No. of Quantity has most sale i.e (20.4%)
####10 No. of Quantity has most discount i.e (8.9%)
####13 No. of Quantity has most profit i.e (27.6%)

##Sub-Catagory wise Analysis of the Profit, Sales and Discount

```python
df_sub_category = df.groupby(['Sub-Category'])[['Sales', 'Discount', 'Profit']].mean()
df_sub_category.head(10)
```

###1.Sales

```python
plt.figure(figsize = (15,10))
plt.pie(df_sub_category['Sales'], labels = df_sub_category.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Sales Analysis', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

###From above
####Top 2 sales sub-category are 
####1.Copier (30.2%)
####2.Machine (22.6)

###2.Profit

```python
plt.figure(figsize = (15,12))
plt.pie(df_sub_category['Profit'], labels = df_sub_category.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Profit Analysis', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

###From above
####Copier has the max profit i.e 74.7%

###3.Discount

```python
plt.figure(figsize = (15,10))
plt.pie(df_sub_category['Discount'], labels = df_sub_category.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Discount Analysis', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

###From above
####Binder and Machine are the top 2 with most discount
